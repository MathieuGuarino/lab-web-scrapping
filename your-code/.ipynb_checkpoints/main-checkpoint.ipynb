####  20 latest earthquakes info (date, time, latitude, longitude and region name) by the EMSC as a pandas dataframe

````python
# This is the url you will scrape in this exercise
url = 'https://www.emsc-csem.org/Earthquake/'
html = requests.get(url).content
soup = BeautifulSoup(html, "lxml")
````

````python
import numpy

colnames = soup.find(id="haut_tableau").find_all('th')
colnames_txt =  [name.text for name in colnames][1:7]
del colnames_txt[2:4]
eq_list = soup.find(id='tbody')

def formatData(soupList):
    rows = (eq_list.find_all('tr',limit=20))
    names = [row.find_all('td')[11].text.strip() for row in rows]
    latitudes = [row.find('td',{'class':'tabev1'}).text.strip('\xa0') for row in rows]
    longitudes = [row.find_all('td',{'class':'tabev1'})[1].text.strip('\xa0') for row in rows]
    dates = ([' '.join(list(filter(None,row.find_all('td',{'class':'tabev6'})[0].text.strip('earthquake').split('\xa0')))) for row in rows])
    return [dates, latitudes, longitudes, names]

df = pd.DataFrame(numpy.array(formatData(eq_list)).T, columns=colnames_txt)
df
````

#### List all language names and number of related articles in the order they appear in wikipedia.org

````python
# This is the url you will scrape in this exercise
url = 'https://www.wikipedia.org/'
html = requests.get(url).content
soup = BeautifulSoup(html, "lxml")
````

````python
mydivs = soup.find_all("div", {"class": "central-featured-lang"})

langages_and_articles = [(''.join(div.find('small').text.split('\xa0')),div.find('strong').text) for div in mydivs]
langages_and_articles
````

#### Top 10 languages by number of native speakers stored in a Pandas Dataframe

````python
# This is the url you will scrape in this exercise
url = 'https://en.wikipedia.org/wiki/List_of_languages_by_number_of_native_speakers'
html = requests.get(url).content
soup = BeautifulSoup(html, "lxml")
````

````python
table = soup.find('table',{'class':'wikitable'})
languages = table.find_all('tr',limit=11)[1:]
top10 = [language.find('a').text for language in languages]
top10
````
